Creating a Dataframe and plotting graphs

```
import pandas as pd
import numpy as np
#Creating dataframe for a table of values
salesMen=["Ahmed","Omar","Ali","Zaid","Salwa","Lila"]
Mobile_sales=[2540,1370,1320,2000,2100,2150]
Tv_sales=[2200,1900,2150,1850,1770,2000]
#stored as list initially
#so converting to dataframe to display as a table
df=pd.DataFrame()
#column names are given
df['Name']=salesMen
df['Mobile_sales']=Mobile_sales
df['TV_Sales']=Tv_sales
#based on name index it is arranged-0 is ahmed and so on
df.set_index("Name",inplace=True)
df
```

```
df.plot.bar(figsize=(20,10))
#Ahmed sold most number of mobiles
#Matplotlib and seaborn-packages for plots
#with the help of dataframe we can draw bar graph
```

```
df.plot.pie(subplots=True)
#seperate for mobile and tv
```

```
df=pd.read_csv('NCHS.csv')
print(df.to_string())
```

```
#tot no of records in dataset
df.shape
#rows and columns (r,c)
```

```
data=df.dropna()
df.shape
#no rows are dropped so no rows have null values
```

```
#unique death causes in US
causes=df["Cause Name"].unique()
causes
```

```
#Remove all causes from the cause name
data=df[df["Cause Name"]!="All causes"]
causes=data["Cause Name"].unique()
causes
```

```
#how many diseases totally
len(causes)
```

```
#unique states in study
state=data["State"].unique()
state
```

```
len(state)
```

```
#total number of deaths in united states
data["Deaths"].sum()
```

```
#total deaths in US
data["Deaths"].sum()
```

```
#no of deaths for each year from 1995-2017
dyear=data.groupby(["Year"]).sum()
dyear
```

```
dyear["Deaths"].plot(title="Deaths per year")
```
